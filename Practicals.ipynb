Task:
Suppose we are given some data for a company sales in a given year,
And we are required to make a prediction of the sales of that company in a given year 'x'


```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Loading the dataset
data = pd.read_csv("C:/Users/user/Desktop/Assignment/Emmerging assignment/Datasets/01-sales_dataset.csv")

#View the data
display(data)

#Exploring the data
print("The shape of the given dataset is: {}".format(data.shape))

#Description of the dataset
print("Description of the dataset\n{}".format(data.describe()))
```

```python
#Ploting the data
data.plot(x = "year", y = "sales", style = "o")
plt.title("YEAR VS SALES")
plt.xticks(np.arange(min(data['year']), max(data['year'])+1, 1.0))
plt.ylabel("Sales")
plt.xlabel("Year")
plt.show()
```

```python
#Splitting the data into training and test sets
from sklearn.model_selection import train_test_split
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

print(X)
print(y)
```

```python
#An alternative way
display(data)
X = np.vstack(data["year"])
print(X)
```

```python
#Split train set and test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Building the LinearRegression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

#Evaluating the derived parameter/Calculated parameter
print(f'b = {model.intercept_}')
print(f'w = {model.coef_}')


```

```python
#An alternative way of formating string
print("b = {}".format(model.intercept_))
print("w = {}".format(model.coef_))
```

```python
#Using the model to make predictions for the test data set
y_pred = model.predict(X_test)
print("X_test are:\n{}\ny_pred are:\n{}".format(X_test, y_pred))
```

```python
#Predict all values of y and plot the data
all_y_pred = model.predict(X)
plt.figure(figsize=(12, 6))
plt.plot(X, all_y_pred) # Regression line
plt.plot(X, y, 'ro') # Scatter plot showing actual data
plt.title('Actual vs Predicted')
plt.xticks(np.arange(min(X), max(X)+1, 1.0))
plt.xlabel('Years')
plt.ylabel('Sales')
```

```python
#Evaluating the model
from sklearn.metrics import mean_squared_error, r2_score

print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```

```python
#There are many ways of evaluating a model
#Let's try to use score() function
print("Training accuracy score: {}".format(model.score(X_train, y_train)))
print("Testing accuracy score: {}".format(model.score(X_test, y_test)))
```

The chain already has restaurants in various cities and you have data for profits and populations from the cities. You would like to use this data to help you select which city to expand to next.

The dataset is contained in the file dataset/01-profits_and_populations.csv. The first column is the population of a city and the second column is the profit of a food truck in that city (a negative value for profit indicates a loss).

Create a model that will project the profits of establishing a restaurant in a given restaurant in a city with a given population to help you in the decision making process.


```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Loading the dataset
data_02 = pd.read_csv("C:/Users/user/Desktop/Assignment/Emmerging assignment/Datasets/02-profits_and_populations.csv")

#View the data
display(data_02)
```

```python
#Exploring the data
print("The shape of the given dataset is: {}".format(data_02.shape))

#Description of the dataset
print("Description of the dataset\n{}".format(data_02.describe()))
```

```python
#Ploting the data
data_02.plot(x = "population", y = "profit", style = "o")
plt.title("POPULATION VS PROFIT")
plt.xticks(np.arange(min(data_02['population']), max(data_02['population']) + 2, 2))
plt.ylabel("population")
plt.xlabel("profit")
plt.show()
```

```python
#Splitting the data into training and test sets
from sklearn.model_selection import train_test_split
X = data_02.iloc[:, :-1].values
y = data_02.iloc[:, 1].values

print(X)
print(y)
```

```python
#Split train set and test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Building the LinearRegression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

#An derived values
print("b = {}".format(model.intercept_))
print("w = {}".format(model.coef_))
```

```python
#Model Evaluation
from sklearn.metrics import mean_squared_error, r2_score
print("Training accuracy score: {}".format(model.score(X_train, y_train)))
print("Testing accuracy score: {}".format(model.score(X_test, y_test)))

y_pred = model.predict(X_test)
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
mean_squared_before = mean_squared_error(y_test, y_pred)
r2_score_before = r2_score(y_test, y_pred)
```

```python
#Model improvement
#Let us improve our model now
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#Let us fit the scaler
scaler.fit(X_train)
#Then Transform the X_train set
X_train_scaled = scaler.transform(X_train)
print("Transformed shape: {}".format(X_train_scaled.shape))
print("Shape before transformation: {}".format(X_train.shape))
```

The MinMaxScaler, shifts the data such that all features are exactly between 0 and 1

```python
#Then we transform the X_test set
X_test_scaled = scaler.transform(X_test)
```

```python
#Then we build a model using a scaled data
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
#now let's evaluate a model
from sklearn.metrics import mean_squared_error, r2_score
y_pred = linear.predict(X_test_scaled)
print("A new R2 score is: {}".format(r2_score(y_test, y_pred)))
print("A new mean squared error is: {}".format(mean_squared_error(y_test, y_pred)))
mean_squared_after = mean_squared_error(y_test, y_pred)
r2_score_after = r2_score(y_test, y_pred)
```

```python
#Comparison of the values before and after scaling the data
print("Mean Squared error before scaling is: {}\nMean Squared error after scaling is: {}".format(mean_squared_before, mean_squared_after))
print("R2 score before scaling is: {}\nR2 score after scaling is: {}".format(r2_score_before, r2_score_after))
```

To develop the best medical insurance products, the insurer needs access to historical data to approximate the medical costs of each user. With this data, a medical insurer can develop more accurate pricing models, plan a particular insurance outcome. For all these cases, the objective is to accurately predict insurance costs.

Dataset
You are provided with a dataset in the dataset/04-insurance_charges.csv file that contains medical insurance records. The individual medical costs billed by health insurance are the target variable charges, and the rest of columns contain personal information such as age, gender, family status, and whether the patient smokes among other features. The following is the description of the features contained in the dataset:

age: age of primary beneficiary
sex: insurance contractor gender, female, male
bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9
children: Number of children covered by health insurance / Number of dependents
smoker: Smoking
region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.
charges: Individual medical costs billed by health insurance

Use Case
The objective is to train a ML regression model that can accurately predict the insurance costs for a given individual



```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#Loading the dataset
#Loading the datasets
data_05 = pd.read_csv("C:/Users/user/Desktop/Assignment/Emmerging assignment/Datasets/05-insurance_charges.csv")

#View the data
display(data_05)
```

```python
#Data description
data_05.describe()
```

```python
#Data Exploration
print("Shape of the dataset is: {}".format(data_05.shape))
#Feature correlation
#We use a heatmap using seaborn library
plt.figure(figsize = (12, 6))
sns.heatmap(data_05.corr(), cmap = 'BrBG', fmt = '.2f', linewidths = 2, annot = True)
```

As we have seen on the heatmap, 3 features which are sex,smoke and region are not shown on the heatmap because are the categorical values. A computer can not understand those categorical values, so we must decode them in order the computer to understand them. There are different methods/ways to deal with those categorical values. One of the method to deal with categorical values is called One-Hot-Encoding(Dummy Variables). This method can be implemented by either using a get_dummies() function which is found in pandas library or that implemented in scikit library.

```python
#Implementation of One-Hot-Encoding method
data_dummies = pd.get_dummies(data_05)
display(data_dummies)
```

```python
print("New shape of the dataset is: {}".format(data_dummies.shape))
```

```python
#Spliting the dataset
X = features = data_dummies.drop("charges", axis = 1)
y = data_dummies.iloc[:, 3]
print("X/features are:")
display(features)
print("y values are:\n{}".format(y))
```

```python
#Spliting the train set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

#Building a linear regression model for multiple features
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train, y_train)
```

```python
#Evaluating the models
print("Train accuracy score: {}".format(linear.score(X_train, y_train)))
print("Test accuracy score: {}".format(linear.score(X_test, y_test)))
```

```python
#Try to use other metrics for evaluating a model
from sklearn.metrics import mean_squared_error, r2_score
y_pred = linear.predict(X_test)
print("Mean squred error: {}".format(mean_squared_error(y_test, y_pred)))
print("R2 score: {}".format(r2_score(y_test,y_pred)))
mean_squared_before = mean_squared_error(y_test, y_pred)
r2_score_before = r2_score(y_test, y_pred)
```

```python
#Model improvement
#Let us improve our model now
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#Let us fit the scaler
scaler.fit(X_train)
#Then Transform the X_train set
X_train_scaled = scaler.transform(X_train)
print("Transformed shape: {}".format(X_train_scaled.shape))
print("Shape before transformation: {}".format(X_train.shape))
```

The MinMaxScaler, shifts the data such that all features are exactly between 0 and 1

```python
#Then we transform the X_test set
X_test_scaled = scaler.transform(X_test)
```

```python
#Then we build a model using a scaled data
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
#now let's evaluate a model
from sklearn.metrics import mean_squared_error, r2_score
y_pred = linear.predict(X_test_scaled)
print("A new R2 score is: {}".format(r2_score(y_test, y_pred)))
print("A new mean squared error is: {}".format(mean_squared_error(y_test, y_pred)))
mean_squared_after = mean_squared_error(y_test, y_pred)
r2_score_after = r2_score(y_test, y_pred)
```

After scaling the data, there is small decrease in mean squared error. Also there is small increase of r2 score.
